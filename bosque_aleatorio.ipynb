```
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification#incluido en sklearn
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import time#valores aleatorios constantes
```

Crear la data del dataset(variables)

```
x, y = make_classification(n_samples=1000, n_features=4,n_informative=2, random_state=0)
```

```
print(x,y)
```

```
print(x.shape)
plt.figure()
plt.title("Dos caracteristicas informativas")
plt.scatter(x[:,0], x[:,1],marker="o",c=y,edgecolors="k")#x variable de caracteristicas y las 2 caractarecteristicas 1 y segunda colomna , y k es una paleta de color diferenciador
plt.show() 
```

Division del conjunto de entrenamiento y conjunto de prueba

```
X_trn,X_tst,Y_trn,Y_tst = train_test_split(x,y,test_size=0.3,random_state=int(time.time()))

print(X_trn.shape)
```

modelado de Random Forest

```
forest=RandomForestClassifier(n_estimators=100, max_depth=2, random_state=0)

forest.fit(X_trn,Y_trn)

print('Precisión en el conjunto de entrenamiento:{:.3f}'.format(forest.score(X_trn,Y_trn)))

print('Precisión en el conjunto de prueba:{:-3f}'.format(forest.score(X_tst,Y_tst)))#debe ser mayor a 85 y con este se valida el modelo
```

Generar una prediccion

```
print(forest.predict([[989898989, 98989899, 9989898989, 848455465]]))
print(forest.predict([[87878788, 87878787, 848445, 8484545]]))
```

matriz de confusion

```
from sklearn.metrics import confusion_matrix

y_true = ["cat", "ant", "cat", "cat", "ant", "bird"] # Clases verdaderas del dataset

y_pred = ["ant", "ant", "cat", "cat", "ant", "cat"] # Clases estimadas por el modelo

confusion_matrix(y_true, y_pred)
```
